```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "area": ["Low pop", "High pop"],
    "waste_before": [2, 200],
    "waste_after":  [4, 220]   # doubling vs +10%
})

print("Raw data:")
print(df)
```

```python
df["abs_change"] = df["waste_after"] - df["waste_before"]

print("\nAbsolute change (raw scale):")
print(df[["area", "waste_before", "waste_after", "abs_change"]])
```

```python
# ----------------------------
# 3) Compute relative changes (% change)
# ----------------------------
df["rel_change"] = df["abs_change"] / df["waste_before"]
df["pct_change"] = 100 * df["rel_change"]

print("\nRelative change (%):")
print(df[["area", "waste_before", "waste_after", "pct_change"]])
```

```python
# ----------------------------
# 4) Take logs (log scale)
#    Use log1p if you might have zeros in real data.
# ----------------------------
df["log_before"] = np.log(df["waste_before"])
df["log_after"]  = np.log(df["waste_after"])

# Difference in logs = log(after) - log(before) = log(after/before)
df["log_diff"] = df["log_after"] - df["log_before"]
df["log_ratio"] = np.log(df["waste_after"] / df["waste_before"])  # same as log_diff

print("\nLog scale changes:")
print(df[["area", "log_before", "log_after", "log_diff", "log_ratio"]])
```

```python
# ----------------------------
# 5) Demonstrate the key property:
#    log(after) - log(before) = log(after/before)
# ----------------------------
print("\nCheck identity (log(after)-log(before) == log(after/before)):")
print(np.allclose(df["log_diff"], df["log_ratio"]))
```
